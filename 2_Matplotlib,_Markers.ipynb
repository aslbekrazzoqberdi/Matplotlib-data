```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3,8,1,10])
```

```python
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
y = np.array([2,4,6,8])
```

```python
plt.plot(y, marker = '*')
plt.show()
```

```python
z = np.array([1,5,2,8])
```

```python
plt.plot(z, marker = '.')
plt.show()
```

This parameter is also called fmt, and is written with this syntax:

```python
ypoints = np.array([3,8,1,10])
```

```python
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
x = np.array([5,15,8,20])
```

```python
plt.plot(x, '*--k')
plt.show()
```

```python
u = np.array([10,50,30,80])
```

```python
plt.plot(u, marker = '*', ms = 20)
plt.show()
```

You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

```python
zpoints = np.array([10,50,20,70,15])
```

```python
plt.plot(zpoints, marker = 'o', ms = 20, mec = 'k')
plt.show()
```

```python
plt.plot(zpoints, marker = '>', ms = 20, mfc = 'r')
plt.show()
```
