```python
import matplotlib.pyplot  as plt
import numpy as np
```

```python
y = np.array([5, 10, 3, 1])
```

```python
plt.plot(y, marker = 'o', ms = 20,  ls = '--', c = 'r')
plt.show()
```

```python
x = np.array([15,60,30,45])
```

```python
plt.plot(x, marker = 'o', c = 'b', linewidth = 15)
plt.show()
```

```python
z = np.array([15,35,25,45])
```

```python
plt.plot(z, c = 'y', linewidth = 25)
plt.show()
```

```python
a = np.array([10,25,15,50])
b = np.array([25,10,40,45])
```

```python
plt.plot(a, ls = '--', c = 'y')
plt.plot(b, ls = '--', c = 'b')
plt.show()
```
