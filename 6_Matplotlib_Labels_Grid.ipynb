With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([5,10,15,20])
y = np.array([120,130,150,175])
```

```python
plt.plot(x,y, marker = 'o', mfc = 'r', ms = 15)
plt.xlabel("Age of boy")
plt.ylabel("Height of boy")
plt.title("Procces of young boys ")


plt.show()
```
